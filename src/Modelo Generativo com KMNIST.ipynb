```python
# Importar bibliotecas
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras import layers
import os

def load_mnist_images(filepath):
    # Lê o arquivo de imagens
    with open(filepath, 'rb') as f:
        f.read(16)  # Pular os primeiros 16 bytes
        images = np.frombuffer(f.read(), dtype=np.uint8)
    return images.reshape(-1, 28, 28)  # Redimensionar

def load_mnist_labels(filepath):
    # Lê o arquivo de labels
    with open(filepath, 'rb') as f:
        f.read(8)  # Pular os primeiros 8 bytes
        labels = np.frombuffer(f.read(), dtype=np.uint8)
    return labels

# Definir os caminhos para os arquivos extraídos
train_images_path = "train-images-idx3-ubyte"
train_labels_path = "train-labels-idx1-ubyte"
test_images_path = "t10k-images-idx3-ubyte"
test_labels_path = "t10k-labels-idx1-ubyte"

# Carregar os arquivos
x_train = load_mnist_images(train_images_path)
y_train = load_mnist_labels(train_labels_path)
x_test = load_mnist_images(test_images_path)
y_test = load_mnist_labels(test_labels_path)

# Normalizar os dados entre -1 e 1
x_train = (x_train.astype('float32') - 127.5) / 127.5
x_test = (x_test.astype('float32') - 127.5) / 127.5

# Expandir a dimensão dos dados
x_train = np.expand_dims(x_train, axis=-1)
x_test = np.expand_dims(x_test, axis=-1)

# Verificar as dimensões
print(f"x_train shape: {x_train.shape}, y_train shape: {y_train.shape}")
print(f"x_test shape: {x_test.shape}, y_test shape: {y_test.shape}")
```

```python
# Juntar o conjunto de treino e teste
dataset = np.concatenate([x_train, x_test], axis=0)

BATCH_SIZE = 64

# Criar o dataset do TensorFlow
dataset = tf.data.Dataset.from_tensor_slices(dataset)
dataset = dataset.shuffle(buffer_size=1024).batch(BATCH_SIZE)

# Verificar o batch de dados
for batch in dataset.take(1):
    print(batch.shape)
```

```python
# Dimensão do ruído
NOISE_DIM = 100

# Modelo Gerador
generator = tf.keras.Sequential([
    layers.Dense(7*7*256, use_bias=False, input_shape=(NOISE_DIM,)),
    layers.BatchNormalization(),
    layers.LeakyReLU(),
    
    layers.Reshape((7, 7, 256)),
    
    layers.Conv2DTranspose(128, kernel_size=5, strides=1, padding="same", use_bias=False),
    layers.BatchNormalization(),
    layers.LeakyReLU(),
    
    layers.Conv2DTranspose(64, kernel_size=5, strides=2, padding="same", use_bias=False),
    layers.BatchNormalization(),
    layers.LeakyReLU(),
    
    layers.Conv2DTranspose(1, kernel_size=5, strides=2, padding="same", use_bias=False, activation='tanh')
])

# Exibir o resumo do gerador
generator.summary()
```

```python
# Modelo Discriminador
discriminator = tf.keras.Sequential([
    layers.Conv2D(64, kernel_size=5, strides=2, padding="same", input_shape=[28, 28, 1]),
    layers.LeakyReLU(),
    layers.Dropout(0.3),
    
    layers.Conv2D(128, kernel_size=5, strides=2, padding="same"),
    layers.LeakyReLU(),
    layers.Dropout(0.3),
    
    layers.Flatten(),
    layers.Dense(1, activation='sigmoid')
])

# Exibir o resumo do discriminador
discriminator.summary()
```

```python
# Função de perda
cross_entropy = tf.keras.losses.BinaryCrossentropy(from_logits=True)

# Otimizadores
generator_optimizer = tf.keras.optimizers.Adam(1e-5)
discriminator_optimizer = tf.keras.optimizers.Adam(1e-5)

# Função de perda do discriminador
def discriminator_loss(real_output, fake_output):
    real_loss = cross_entropy(tf.ones_like(real_output), real_output)
    fake_loss = cross_entropy(tf.zeros_like(fake_output), fake_output)
    return real_loss + fake_loss

# Função de perda do gerador
def generator_loss(fake_output):
    return cross_entropy(tf.ones_like(fake_output), fake_output)
```

```python
# Função de treino
@tf.function
def train_step(images):
    noise = tf.random.normal([BATCH_SIZE, NOISE_DIM])

    with tf.GradientTape() as gen_tape, tf.GradientTape() as disc_tape:
        generated_images = generator(noise, training=True)

        real_output = discriminator(images, training=True)
        fake_output = discriminator(generated_images, training=True)

        gen_loss = generator_loss(fake_output)
        disc_loss = discriminator_loss(real_output, fake_output)

    gradients_of_generator = gen_tape.gradient(gen_loss, generator.trainable_variables)
    gradients_of_discriminator = disc_tape.gradient(disc_loss, discriminator.trainable_variables)

    generator_optimizer.apply_gradients(zip(gradients_of_generator, generator.trainable_variables))
    discriminator_optimizer.apply_gradients(zip(gradients_of_discriminator, discriminator.trainable_variables))
```

```python
# Função para salvar as imagens geradas
def generate_and_save_images(model, epoch):
    noise = tf.random.normal([16, NOISE_DIM])
    generated_images = model(noise, training=False)

    fig = plt.figure(figsize=(4, 4))

    for i in range(generated_images.shape[0]):
        plt.subplot(4, 4, i + 1)
        plt.imshow(generated_images[i, :, :, 0] * 127.5 + 127.5, cmap='gray')
        plt.axis('off')

    plt.savefig(f"image_at_epoch_{epoch}.png")
    plt.show()
```

```python
# Função para treinar o modelo
def train(dataset, epochs):
    for epoch in range(epochs):
        for image_batch in dataset:
            train_step(image_batch)

        # Gerar e mostrar amostra de imagens a cada 10 epochs
        if (epoch + 1) % 10 == 0:
            generate_and_save_images(generator, epoch + 1)
```

```python
# Definir número de épocas
EPOCHS = 250

# Treinar o modelo' 
train(dataset, EPOCHS)
```
